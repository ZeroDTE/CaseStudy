````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlations between numeric features
numeric_features = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 
                   'crispedricewafer', 'hard', 'bar', 'pluribus', 'sugarpercent', 
                   'pricepercent', 'winpercent']
correlation_matrix = data_copy[numeric_features].corr()

# Create correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, 
            annot=True,
            cmap='coolwarm',
            center=0,
            fmt='.2f',
            square=True)
plt.title('Correlation Matrix of Candy Features', pad=20)
plt.tight_layout()
plt.show()

# Feature correlations with win percentage (horizontal bar plot)
correlations = correlation_matrix['winpercent'].sort_values(ascending=True)
correlations = correlations.drop('winpercent')  # Remove self-correlation

# Create a DataFrame for seaborn to plot
corr_df = pd.DataFrame({'features': correlations.index, 'correlation': correlations.values})
plt.figure(figsize=(10, 6))
sns.barplot(data=corr_df, 
           y='features', 
           x='correlation',
           palette=['red' if x < 0 else 'green' for x in correlations.values])
plt.title('Feature Correlations with Win Percentage')
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Calculate and display summary statistics
print("\nCorrelations with Win Percentage:")
for feature, corr in correlations.items():
    print(f"{feature}: {corr:.3f}")


plt.show()
````

````python
# Import scipy.stats for statistical testing
from scipy import stats

# Analyze win percentage difference for all features
features_to_analyze = correlation_matrix.columns.drop('winpercent').tolist()
fig, axes = plt.subplots(len(features_to_analyze), 2, figsize=(15, 5*len(features_to_analyze)))

for idx, feature in enumerate(features_to_analyze):
    with_feature = data_copy[data_copy[feature] == 1]['winpercent'] if feature not in ['sugarpercent', 'pricepercent'] else \
                   data_copy[data_copy[feature] > data_copy[feature].median()]['winpercent']
    without_feature = data_copy[data_copy[feature] == 0]['winpercent'] if feature not in ['sugarpercent', 'pricepercent'] else \
                      data_copy[data_copy[feature] <= data_copy[feature].median()]['winpercent']
    
    feature_label = feature.replace('percent', ' %').title()
    
    # Bar chart showing averages
    avg_data = [with_feature.mean(), without_feature.mean()]
    bars = axes[idx, 0].bar([f'High {feature_label}' if feature in ['sugarpercent', 'pricepercent'] else f'With {feature_label}',
                            f'Low {feature_label}' if feature in ['sugarpercent', 'pricepercent'] else f'Without {feature_label}'],
                           avg_data,
                           color=['#2ecc71', '#e74c3c'],
                           width=0.6)
    
    # Add value labels on bars
    for bar in bars:
        height = bar.get_height()
        axes[idx, 0].text(bar.get_x() + bar.get_width()/2., height,
                         f'{height:.1f}%',
                         ha='center', va='bottom')
    
    axes[idx, 0].set_title(f'Average Win % Comparison - {feature_label}', pad=40, fontsize=12)
    axes[idx, 0].set_ylabel('Win Percentage')
    axes[idx, 0].grid(axis='y', linestyle='--', alpha=0.3)
    
    # Add statistical significance annotation
    t_stat, p_value = stats.ttest_ind(with_feature, without_feature)
    sig_text = "Statistically Significant" if p_value < 0.05 else "Not Statistically Significant"
    axes[idx, 0].text(0.5, max(avg_data) + 5,
                     f'Difference: {abs(avg_data[0] - avg_data[1]):.1f}%\n({sig_text})',
                     ha='center', va='bottom',
                     bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray'))
    
    # Distribution plot
    violin_parts = axes[idx, 1].violinplot([with_feature, without_feature],
                                         showmeans=True, showmedians=True)
    
    # Customize violin plot colors
    violin_parts['bodies'][0].set_facecolor('#2ecc71')
    violin_parts['bodies'][1].set_facecolor('#e74c3c')
    
    axes[idx, 1].set_xticks([1, 2])
    axes[idx, 1].set_xticklabels([f'High {feature_label}' if feature in ['sugarpercent', 'pricepercent'] else f'With {feature_label}',
                                 f'Low {feature_label}' if feature in ['sugarpercent', 'pricepercent'] else f'Without {feature_label}'])
    axes[idx, 1].set_title(f'Win % Distribution - {feature_label}', pad=40, fontsize=12)
    axes[idx, 1].set_ylabel('Win Percentage')
    axes[idx, 1].grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

def prepare_data(data, test_size=0.2, val_size=0.2, random_state=42):
    features = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 
                'nougat', 'crispedricewafer', 'sugarpercent', 'pricepercent']
    
    X = data[features].values
    y = data['winpercent'].values
    
    # Add standardization here
    X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
    
    # Splits remain the same
    X_temp, X_test, y_temp, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )
    
    val_size_adjusted = val_size / (1 - test_size)
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=val_size_adjusted, random_state=random_state
    )
    
    return X_train, X_val, X_test, y_train, y_val, y_test

def gradient_descent(X, y, learning_rate=0.001, lambda_reg=1.0, max_iterations=10000, epsilon=1e-6):
    m, n = X.shape  # m samples, n features
    
    # Create polynomial features (simplified version)
    X_poly = np.column_stack([
        X,  # linear terms
        # Only include meaningful interactions
        np.array([X[:,i] * X[:,j] 
                 for i, j in [(0,3), (0,4), (0,5), (6,7)]])  # chocolatepeanut, chocolatenougat, chocolatecrisp, sugarprice
        .T
    ])
    
    # Add column of ones for intercept term
    X_b = np.c_[np.ones((m, 1)), X_poly]
    
    # Initialize parameters (betas)
    betas = np.random.randn(X_b.shape[1])
    
    # Store cost history for plotting
    cost_history = []
    
    for iteration in range(max_iterations):
        # Compute predictions
        y_pred = X_b.dot(betas)
        
        # Compute gradients (with regularization, excluding intercept)
        gradients = (2/m) * X_b.T.dot(y_pred - y)
        gradients[1:] += (2 * lambda_reg * betas[1:]) / m  # Add regularization term
        
        # Update parameters
        betas_prev = betas.copy()
        betas = betas - learning_rate * gradients
        
        # Compute cost
        cost = (1/m) * np.sum((y_pred - y)**2) + lambda_reg * np.sum(betas[1:]**2)
        cost_history.append(cost)
        
        # Check convergence
        if np.all(np.abs(betas - betas_prev) < epsilon):
            print(f"Converged after {iteration} iterations")
            break
            
    return betas, cost_history

def evaluate_model(X, y, betas):
    """Ensure this matches the feature creation in gradient_descent"""
    # Create polynomial features (same as in gradient_descent)
    X_poly = np.column_stack([
        X,  # linear terms
        # Only include meaningful interactions
        np.array([X[:,i] * X[:,j] 
                 for i, j in [(0,3), (0,4), (0,5), (6,7)]])  # chocolatepeanut, chocolatenougat, chocolatecrisp, sugarprice
        .T
    ])
    
    X_b = np.c_[np.ones((X.shape[0], 1)), X_poly]
    y_pred = X_b.dot(betas)
    mse = np.mean((y_pred - y)**2)
    r2 = 1 - (np.sum((y - y_pred)**2) / np.sum((y - np.mean(y))**2))
    return mse, r2, y_pred

def run_candy_analysis(data):
    # Prepare data with splits
    X_train, X_val, X_test, y_train, y_val, y_test = prepare_data(data)
    
    # Run gradient descent on training data
    betas, cost_history = gradient_descent(X_train, y_train)
    
    # Evaluate model on all splits
    train_mse, train_r2, y_train_pred = evaluate_model(X_train, y_train, betas)
    val_mse, val_r2, y_val_pred = evaluate_model(X_val, y_val, betas)
    test_mse, test_r2, y_test_pred = evaluate_model(X_test, y_test, betas)
    
    # Create feature names (matching the actual features used)
    base_features = ['intercept', 'chocolate', 'fruity', 'caramel', 'peanutyalmondy', 
                    'nougat', 'crispedricewafer', 'sugarpercent', 'pricepercent']
    
    interaction_features = [
        'chocolate_peanutyalmondy',
        'chocolate_nougat',
        'chocolate_crispedricewafer', 
        'sugarpercent_pricepercent'
    ]
            
    feature_names = base_features + interaction_features
    
    coefficients = dict(zip(feature_names[:len(betas)], betas))
    
    return {
        'coefficients': coefficients,
        'train_metrics': {'mse': train_mse, 'r2': train_r2},
        'val_metrics': {'mse': val_mse, 'r2': val_r2},
        'test_metrics': {'mse': test_mse, 'r2': test_r2},
        'predictions': {
            'train': (y_train, y_train_pred),
            'val': (y_val, y_val_pred),
            'test': (y_test, y_test_pred)
        },
        'cost_history': cost_history
    }

# 5. Visualization
def plot_results(results):
    # Plot cost history
    plt.figure(figsize=(10, 5))
    plt.plot(results['cost_history'])
    plt.title('Cost Function over Iterations')
    plt.xlabel('Iteration')
    plt.ylabel('Cost')
    plt.show()
    
    # Plot regression results for all splits
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
    
    for ax, (split, title) in zip(
        [ax1, ax2, ax3],
        [('train', 'Training'), ('val', 'Validation'), ('test', 'Test')]
    ):
        y_true, y_pred = results['predictions'][split]
        ax.scatter(y_true, y_pred, alpha=0.5)
        ax.plot([min(y_true), max(y_true)], 
                [min(y_true), max(y_true)], 
                'r--', lw=2)
        ax.set_xlabel('Actual Win Percentage')
        ax.set_ylabel('Predicted Win Percentage')
        ax.set_title(f'{title} Set Results')
    
    plt.tight_layout()
    plt.show()
    
    # Plot feature importance
    coeffs = results['coefficients']
    plt.figure(figsize=(15, 8))
    plt.bar(coeffs.keys(), coeffs.values())
    plt.xticks(rotation=90, ha='right')
    plt.title('Feature Coefficients (Beta Values)')
    plt.margins(x=0.01)
    plt.tight_layout(pad=1.5)
    plt.show()

def generate_synthetic_candy(model_coefficients, n_samples=1, realistic_constraints=True):
    """
    Generate synthetic candy features with realistic constraints
    """
    feature_ranges = {
        'chocolate': (0, 1),
        'fruity': (0, 1), 
        'caramel': (0, 1),
        'peanutyalmondy': (0, 1),
        'nougat': (0, 1),
        'crispedricewafer': (0, 1),
        'sugarpercent': (0, 1),
        'pricepercent': (0, 1)
    }
    
    synthetic_candies = []
    
    for _ in range(n_samples):
        valid_candy = False
        while not valid_candy:
            features = {}
            
            # Generate binary features
            binary_features = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer']
            for feature in binary_features:
                features[feature] = np.random.choice([0, 1])
                
            # Generate continuous features
            features['sugarpercent'] = np.random.uniform(0.2, 1)  # Minimum sugar content of 0.2
            features['pricepercent'] = np.random.uniform(0.1, 0.8)  # Reasonable price range
            
            if realistic_constraints:
                # Check all constraints
                valid_candy = True
                
                # Chocolate and fruity are mutually exclusive
                if features['chocolate'] == 1 and features['fruity'] == 1:
                    valid_candy = False
                    continue
                    
                # Nougat and crispy wafer require chocolate
                if (features['nougat'] == 1 or features['crispedricewafer'] == 1) and features['chocolate'] == 0:
                    valid_candy = False
                    continue
                    
                # Caramel usually appears with chocolate
                if features['caramel'] == 1 and features['chocolate'] == 0:
                    if np.random.random() < 0.8:  # 80% chance to reject
                        valid_candy = False
                        continue
                        
                # Nuts are more common with chocolate
                if features['peanutyalmondy'] == 1 and features['chocolate'] == 0:
                    if np.random.random() < 0.7:  # 70% chance to reject
                        valid_candy = False
                        continue
                
                # Adjust price based on features
                base_price = features['pricepercent']
                if features['chocolate'] == 1:
                    base_price *= 1.2  # Chocolate candies tend to be pricier
                if features['nougat'] == 1 or features['caramel'] == 1:
                    base_price *= 1.1  # Complex candies cost more
                features['pricepercent'] = min(0.8, base_price)  # Cap at 0.8
                
            if valid_candy:
                # Create feature array for prediction
                X = np.array([[
                    features['chocolate'], features['fruity'], features['caramel'],
                    features['peanutyalmondy'], features['nougat'], features['crispedricewafer'],
                    features['sugarpercent'], features['pricepercent']
                ]])
                
                # Create interaction terms
                X_poly = np.column_stack([
                    X,
                    X[:,0] * X[:,3],  # chocolatepeanut
                    X[:,0] * X[:,4],  # chocolatenougat
                    X[:,0] * X[:,5],  # chocolatecrisp
                    X[:,6] * X[:,7]   # sugarprice
                ])
                
                # Add bias term
                X_b = np.c_[np.ones((1, 1)), X_poly]
                
                # Predict win percentage
                features['predicted_winpercent'] = X_b.dot(list(model_coefficients.values()))[0]
                
                synthetic_candies.append(features)
    
    return pd.DataFrame(synthetic_candies)

# Example usage
if __name__ == "__main__":
    
    results = run_candy_analysis(data_copy)
    plot_results(results)
    
    print("\nModel Performance:")
    print("Training:")
    print(f"MSE: {results['train_metrics']['mse']:.4f}")
    print(f"R: {results['train_metrics']['r2']:.4f}")
    print("\nValidation:")
    print(f"MSE: {results['val_metrics']['mse']:.4f}")
    print(f"R: {results['val_metrics']['r2']:.4f}")
    print("\nTest:")
    print(f"MSE: {results['test_metrics']['mse']:.4f}")
    print(f"R: {results['test_metrics']['r2']:.4f}")
    
    # Generate and evaluate synthetic candies
    new_candies = generate_synthetic_candy(results['coefficients'], n_samples=5)
    print("\nSynthetic Candies:")
    print(new_candies)
````

````python
def print_interesting_coefficients(coefficients, threshold=0.1):
    """
    Print coefficients in a formatted way, highlighting the most influential ones
    
    Args:
        coefficients: Dictionary of feature names and their coefficients
        threshold: Minimum absolute value to consider a coefficient "interesting"
    """
    # Sort coefficients by absolute value
    sorted_coeffs = sorted(
        coefficients.items(), 
        key=lambda x: abs(x[1]), 
        reverse=True
    )
    
    print("Most Influential Features:")
    print("-" * 50)
    print(f"{'Feature':<25} {'Coefficient':>12} {'Impact':<8}")
    print("-" * 50)
    
    for feature, coeff in sorted_coeffs:
        if abs(coeff) >= threshold:
            impact = "" if coeff > 0 else ""
            print(f"{feature:<25} {coeff:>12.3f} {impact:<8}")

# Use it with your existing results
results = run_candy_analysis(data_copy)
print_interesting_coefficients(results['coefficients'])
````

# Sigkeiten-Analyse Dokumentation

## 1. Erweiterter Feature-Vektor

Fr jede Sigkeit i enthlt unser Feature-Vektor x_i die folgenden Komponenten:

$$x_i = \begin{bmatrix} 
1 \\
\text{Basis-Features} \\
chocolate_i \\
fruity_i \\
caramel_i \\
peanutyalmondy_i \\
nougat_i \\
crispedricewafer_i \\
sugarpercent_i \\
pricepercent_i \\
\text{Wechselwirkungsterme} \\
chocolate_i \times nougat_i \\
chocolate_i \times crispedricewafer_i \\
sugarpercent_i \times pricepercent_i
\end{bmatrix}$$

## 2. Vorhersagemodell

Das finale Regressionsmodell enthlt lineare Terme und ausgewhlte Wechselwirkungen:

$$\hat{y}_i = \beta_0 + \sum_{j=1}^p \beta_j x_{ij} + \sum_{selected} \beta_{jk} x_{ij}x_{ik}$$

Die Wechselwirkungsterme wurden basierend auf Domain-Wissen ausgewhlt:
- Schokolade  Nougat
- Schokolade  Knusprige Waffel
- Zuckergehalt  Preis
